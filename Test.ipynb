```python
f1=150
f2=80
f3=40

t=np.linspace(0,1,1000)

#Señal 1
w1=2*np.pi*f1*t
y1=np.cos(w1)

plt.plot(t,y1)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal 1')
plt.xlim(0,0.1)
plt.show()

#Señal 2
w2=2*np.pi*f2*t
y2=np.cos(w2)

plt.plot(t,y2)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal 2')
plt.xlim(0,0.1)
plt.show()

#Señal 3
w3=2*np.pi*f3*t
y3=np.cos(w3)

plt.plot(t,y3)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal 3')
plt.xlim(0,0.1)
plt.show()
```
