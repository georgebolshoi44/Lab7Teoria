```python
import IPython 
import numpy as np
from scipy import signal
from scipy.io import wavfile
import matplotlib.pyplot as plt
```

```python
Fs, x = wavfile.read("vocala.wav")
IPython.display.Audio(x, rate=Fs)
print(Fs)
```

```python
f1=150
f2=80
f3=40

t=np.linspace(0,1,1000)

#Seal 1
w1=2*np.pi*f1*t
y1=np.cos(w1)

plt.plot(t,y1)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Seal 1')
plt.xlim(0,0.1)
plt.show()

#Seal 2
w2=2*np.pi*f2*t
y2=np.cos(w2)

plt.plot(t,y2)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Seal 2')
plt.xlim(0,0.1)
plt.show()

#Seal 3
w3=2*np.pi*f3*t
y3=np.cos(w3)

plt.plot(t,y3)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Seal 3')
plt.xlim(0,0.1)
plt.show()
```

```python
y4 = y1 + y2 +y3

plt.plot(t,y4)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Resultante')
plt.xlim(0,0.1)
plt.show()
```

```python
fourierTransform = np.fft.fft(y4)/len(y4)
fourierTransform = fourierTransform[range(int(len(y4)/2))]

#

tpCount     = len(y4)
values      = np.arange(int(tpCount/2))
print(Fs)
timePeriod  = tpCount / Fs
frequencies = values / timePeriod
plt.plot(frequencies, abs(fourierTransform))
plt.xlim(0,3000)
```
