# PROBLEMA 2

```python
f1=320
f2=220
f3=20

t=np.linspace(0,1,1000)

#Señal 1
w1=2*np.pi*f1*t
y1=np.cos(w1)

plt.plot(t,y1)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal 1')
plt.xlim(0,0.1)
plt.show()

#Señal 2
w2=2*np.pi*f2*t
y2=np.cos(w2)

plt.plot(t,y2)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal 2')
plt.xlim(0,0.1)
plt.show()

#Señal 3
w3=2*np.pi*f3*t
y3=np.cos(w3)

plt.plot(t,y3)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Señal 3')
plt.xlim(0,0.1)
plt.show()
```

```python
y4 = y1 + y2 +y3

plt.plot(t,y4)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('Resultante')
plt.xlim(0,0.1)
plt.show()
```

```python
fourierTransform = np.fft.fft(y4)/len(y4)
fourierTransform = fourierTransform[range(int(len(y4)/2))]

Fs = 1000

tpCount     = len(y4)
values      = np.arange(int(tpCount/2))

timePeriod  = tpCount / Fs
frequencies = values / timePeriod
plt.plot(frequencies, abs(fourierTransform))
plt.xlim(0,500)
```

```python
n = 5

fcl = 150
fch = 280
fcb = np.array([180,400])

w_cl = 2*fcl/Fs
w_ch = 2*fch/Fs
w_cb = 2*fcb/Fs

bfl = 2*pi*fcl
bfh = 2*pi*fch
bfb = 2*pi*fcb
```

```python
#Filtro IIR

[bl,al] = signal.butter(n,w_cl,btype='lowpass')

[wl,hl] = signal.freqz(bl,al,worN=2000)
wl2 = Fs*wl/(2*pi)
hl_db = 20*np.log10(abs(hl))



[bh,ah] = signal.butter(n,w_ch,btype='highpass')

[wh,hh] = signal.freqz(bh,ah,worN=2000)
wh2 = Fs*wh/(2*pi)
hh_db = 20*np.log10(abs(hh))


[bb,ab] = signal.butter(n,w_cb,btype='bandpass')

[wb,hb] = signal.freqz(bb,ab,worN=2000)
wb2 = Fs*wb/(2*pi)
hb_db = 20*np.log10(abs(hb))

plt.plot(wl2,hl_db)
plt.xlabel('Freq')
plt.ylabel('magnitud')
plt.title('Resultante')
plt.show()

plt.plot(wh2,hh_db)
plt.xlabel('Freq')
plt.ylabel('magnitud')
plt.title('Resultante')
plt.show()

plt.plot(wb2,hb_db)
plt.xlabel('Freq')
plt.ylabel('magnitud')
plt.title('Resultante')
plt.show()
```

```python
# Filtro FIR

Nf = 31

tl = signal.firwin(Nf, w_cl,pass_zero='lowpass')
[wfl,hfl] = signal.freqz(tl, worN=2000)
wfl = Fs*wfl/(2*pi)
hfl_db = 20*np.log10(abs(hfl))

th = signal.firwin(Nf, w_ch,pass_zero='highpass')
[wfh,hfh] = signal.freqz(th, worN=2000)
wfh = Fs*wfh/(2*pi)
hfh_db = 20*np.log10(abs(hfh))

tb = signal.firwin(Nf, w_cb,pass_zero='bandpass')
[wfb,hfb] = signal.freqz(tb, worN=2000)
wfb = Fs*wfb/(2*pi)
hfb_db = 20*np.log10(abs(hfb))

plt.plot(wfl,hfl_db)
plt.xlabel('Freq')
plt.ylabel('magnitud')
plt.title('Resultante')
plt.show()

plt.plot(wfh,hfh_db)
plt.xlabel('Freq')
plt.ylabel('magnitud')
plt.title('Resultante')
plt.show()

plt.plot(wfb,hfb_db)
plt.xlabel('Freq')
plt.ylabel('magnitud')
plt.title('Resultante')
plt.show()
```

De los filtros generados previamente 

```python
zi = signal.lfilter_zi(bl, al)
zl, _ = signal.lfilter(bl, al, y4, zi=zi*y4[0])
zl2, _ = signal.lfilter(bl, al, zl, zi=zi*zl[0])
si1 = signal.filtfilt(bl, al, y4)

plt.plot(t,si1)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('S1 filtro IIR')
plt.xlim(0,0.1)
plt.show()



zi = signal.lfilter_zi(bb, ab)
zb, _ = signal.lfilter(bb, ab, y4, zi=zi*y4[0])
zb2, _ = signal.lfilter(bb, ab, zb, zi=zi*zb[0])
si2 = signal.filtfilt(bb, ab, y4)

plt.plot(t,si2)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('S2 Filtro IIR')
plt.xlim(0,0.1)
plt.show()



zi = signal.lfilter_zi(bh, ah)
zh, _ = signal.lfilter(bh, ah, y4, zi=zi*y4[0])
zh2, _ = signal.lfilter(bh, ah, zh, zi=zi*zh[0])
si3 = signal.filtfilt(bh, ah, y4)

plt.plot(t,si3)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('S3 Filtro IIR')
plt.xlim(0,0.1)
plt.show()
```

```python
sf1 = signal.lfilter(tl, 1.0, y4)

sf2 = signal.lfilter(tb, 1.0, y4)
 
sf3 = signal.lfilter(th, 1.0, y4)

plt.plot(t,sf1)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('S1 filtro FIR')
plt.xlim(0,0.1)
plt.show()

plt.plot(t,sf2)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('S2 filtro FIR')
plt.xlim(0,0.1)
plt.show()

plt.plot(t,sf3)
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.title('S3 filtro FIR')
plt.xlim(0,0.1)
plt.show()
```

## Resultados
Ambos filtros lograron recuperar las señales originales que componian a y4, cabe resaltar como el filtro FIR toma un tiempo en ajustarse antes de poder mostrar la señal filtrada, mientras que por su parte el filtro IIR casi que de inmediato procede a mostrar la señal. Tambien se puede ver que en ambos filtros la segunda señal, obtenida mediante un pasabanda, fue la que se vio más afectada por el proceso de filtrado, siendo la más diferente a su señal original, pero aun asi teniendo las caracteristicas de la original

## Guía Rápida
En primer lugar lafrecuencia de muestreo (Fs) debe elegirse de modo que sea por lo menos 2 veces mayor que la maxima frecuencia a medir.
<br>Para el desarrollo de los filtros en el caso del FIR se trabajo con el diseño de ventana, para esto se emplea la funcion "firwin" de la biblioteca signal. A esta función se le debe indicar en primer lugar el orden del filtro, este debe ser un numero impar en caso de que se cree un filtro pasabanda que incluya la frecuencia de nyquist. Seguidamente se indica la frecuencia de corte, solo sera un valor para filtros pasa-bajas o pasa-altas y un array con dos valores para pasa-bandas o tapa-bandas. Finalmente se debe indicar a la funcion que tipo de filtro se desea crear (pasabajas, pasaaltas, tapabanda o pasabanda)
<br> El filtro IIR por otra parte es generado mediante la funcion "butter" de la misma biblioteca que permite generar un filtro tipo butterworth, a esta función los parametros que se le deben indicar son en primer lugar el orden del filtro, seguido por la frecuencia de corte, y por ultimo el tipo de filtro a realizar(una estrucutra muy parecida a la del firwin).
<br> Con los filtros creados se emplea "lfilter" para procesar la señal que se desea filtrar. En el caso del filtro FIR obtenido mediante "firwin" solo se debe introducir el resultado de "firwin", junto con la señal a filtrar para obtener el resultado. Pero para el filtro IIR se requiere un proceso de adaptado de la señal previo mediante "lfilter_zi" y "lfilter" para finalmente mostrar el resultado con la funcion "filtfilt"
